# Bike Demand Modeling
## Notebook 1: Data Description and EDA

```python
import sys
import os
sys.path.insert(0, os.path.abspath('..'))

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from src.preprocess import load_and_prepare
```

```python
DATA_PATH = '../data/raw/day.csv'
day, day_cv = load_and_prepare(DATA_PATH)
day.head()
```

### Data Description

```python
day.info()
```

```python
var_summary = day[["cnt", "temp", "atemp", "hum", "windspeed"]].describe().round(3)
var_summary
```

### Distribution of Response Variable

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

sns.histplot(day["cnt"], bins=30, kde=True, ax=axes[0])
axes[0].set_title("Distribution of Daily Rentals (cnt)")

sns.histplot(day["log_cnt"], bins=30, kde=True, ax=axes[1])
axes[1].set_title("Distribution of Log Daily Rentals")

plt.tight_layout()
plt.show()
```

### Time Patterns

```python
plt.figure(figsize=(12, 4))
plt.plot(pd.to_datetime(day["dteday"]), day["cnt"])
plt.title("Daily Bike Rentals Over Time")
plt.xlabel("Date")
plt.ylabel("Daily rentals (cnt)")
plt.show()
```

```python
plt.figure(figsize=(5, 4))
sns.boxplot(data=day, x="yr", y="cnt")
plt.title("Daily Bike Rentals by Year")
plt.xlabel("Year (0 = 2011, 1 = 2012)")
plt.ylabel("Daily rentals (cnt)")
plt.tight_layout()
plt.show()
```

### Continuous Predictors vs Rentals

```python
plt.figure(figsize=(7, 4.8))
sns.scatterplot(data=day, x="temp", y="cnt", hue="season_label", alpha=0.7)
sns.regplot(data=day, x="temp", y="cnt", scatter=False, lowess=True, color="black")
plt.title("Daily Rentals vs Temperature")
plt.xlabel("Normalized temperature")
plt.ylabel("Daily rentals (cnt)")
plt.legend(title="Season", bbox_to_anchor=(1.02, 1), loc="upper left")
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4.5))

sns.scatterplot(data=day, x="hum", y="cnt", alpha=0.65, ax=axes[0])
sns.regplot(data=day, x="hum", y="cnt", scatter=False, lowess=True, ax=axes[0], color="black")
axes[0].set_title("Daily Rentals vs Humidity")
axes[0].set_xlabel("Normalized humidity")
axes[0].set_ylabel("Daily rentals (cnt)")

sns.scatterplot(data=day, x="windspeed", y="cnt", alpha=0.65, ax=axes[1])
sns.regplot(data=day, x="windspeed", y="cnt", scatter=False, lowess=True, ax=axes[1], color="black")
axes[1].set_title("Daily Rentals vs Wind Speed")
axes[1].set_xlabel("Normalized wind speed")
axes[1].set_ylabel("Daily rentals (cnt)")

plt.tight_layout()
plt.show()
```

### Categorical Predictors vs Rentals

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

sns.boxplot(data=day, x="season", y="cnt", ax=axes[0])
axes[0].set_title("Bike Rentals by Season")
axes[0].set_xlabel("Season")
axes[0].set_ylabel("Daily rentals")

sns.boxplot(data=day, x="workingday", y="cnt", ax=axes[1])
axes[1].set_title("Bike Rentals by Working Day")
axes[1].set_xlabel("Working Day")
axes[1].set_ylabel("")

sns.boxplot(data=day, x="weathersit", y="cnt", ax=axes[2])
axes[2].set_title("Bike Rentals by Weather Condition")
axes[2].set_xlabel("Weather Condition")
axes[2].set_ylabel("")

plt.tight_layout()
plt.show()
```

### Correlation Matrix

```python
corr_cols = ["cnt", "log_cnt", "temp", "atemp", "hum", "windspeed"]
plt.figure(figsize=(8, 6))
sns.heatmap(day[corr_cols].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

**Key observations:**
- Possible multicollinearity between temp and atemp
- Humidity and wind speed have a negative correlation with rentals
- Temperature likely has a nonlinear relationship with rentals
